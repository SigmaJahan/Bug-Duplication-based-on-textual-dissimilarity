## Significance test (for Normal Distribution)

```python
# t test for normal distribution (Parametric test)
#Welch’s t-test as sample size same
# Mann-whitney test for not normal distribution (Non -parametric test)
# If p< 0.05, You can also report Cohen's d:
# if d < 0.2: it means that the difference is too weak,
# if 0.2<d<0.5: it means a weak difference
# for 0.5<d<0.8: the difference between two dataset is moderate
# for 0.8<d: the difference is large
```

```python
# For normal distribution of Eclipse_data_sim_LDA,Eclipse_data_dissim_LDA
from scipy import stats

t_value,p_value=stats.ttest_ind(Eclipse_data_sim_LDA,Eclipse_data_dissim_LDA)
print('Test statistic is %f'%float("{:.6f}".format(t_value)))
print('p-value for two test is %f'%p_value)
```

```python
#Cohen suggested that d = 0.2 be considered a 'small' effect size, 0.5 represents a 'medium' effect size and 0.8 a 'large' effect size. 
from statistics import mean, stdev
from math import sqrt
cohens_d = (mean(Eclipse_data_sim_LDA) - mean(Eclipse_data_dissim_LDA)) / (sqrt((stdev(Eclipse_data_sim_LDA) ** 2 + stdev(Eclipse_data_dissim_LDA) ** 2) / 2))
print(cohens_d) 
```
